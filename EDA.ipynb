# Explorary Data Analysis

We will perform EDA on the magnetic field data.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import api as magretraceapi
import scipy.signal as signal
```

```python
api = magretraceapi.MagRetraceDataAPI("./data")
```

```python
api.list_template_titles()
print()
api.list_traversal_titles()
```

```python
traversal_data_1 = api.get_traversal_data(0)
traversal_data_2 = api.get_traversal_data(1)
traversal_data_3 = api.get_traversal_data(4)
```

Exploring if retracing is effective.

```python
pos_corr = np.correlate(traversal_data_1.mag_abs, np.flip(traversal_data_2.mag_abs), mode='full')
neg_corr = np.correlate(traversal_data_1.mag_abs, traversal_data_3.mag_abs, mode='full')
plt.plot(pos_corr)
plt.plot(neg_corr)
plt.legend(["positive", "negative"])
```
